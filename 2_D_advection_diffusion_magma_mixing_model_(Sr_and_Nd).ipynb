```python
"""
Supplementary Python code for a 2-D advection--diffusion magma mixing model
============================================================================

Model purpose
-------------
This script simulates the spatial and temporal evolution of Sr and Nd total
concentrations and their radiogenic isotope ratios during two-dimensional
melt mixing. The model is designed for illustrating non-equilibrium mixing
between two initially juxtaposed melt endmembers. It combines chaotic
advection generated by an alternating sine-flow velocity field with explicit
finite-difference diffusion.

Main outputs
------------
The script saves snapshots and figures showing the evolution of:

- 87Sr/86Sr
- 143Nd/144Nd
- Sr/Nd, used here as a proxy for Sr/Sr* in plotting
- total Sr concentration
- total Nd concentration
- an optional tracer-like endmember fraction field F_A

Optional natural-sample overlay
-------------------------------
If an Excel file named ``MI_WR.xlsx`` is placed in the working directory, the
script attempts to read melt inclusion (MI) and whole-rock (WR) data and
overlay them on selected model plots. The expected data format is described
near the end of the script.

"""

# ====================== 1. Import basic libraries ======================
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams.update({  # Set global font sizes for all figures.
    "font.size": 16,
    "axes.titlesize": 18,
    "axes.labelsize": 18,
    "xtick.labelsize": 10,  # Tick-label font size; increase if needed for publication figures.
    "ytick.labelsize": 10,  # Tick-label font size; increase if needed for publication figures.
    "legend.fontsize": 14,
    "figure.titlesize": 24
})

from matplotlib.ticker import MaxNLocator, FormatStrFormatter
from dataclasses import dataclass, asdict
import csv, os
import pandas as pd

# The code below is intentionally written as a single self-contained script so
# that it can be run directly in Python, Google Colab, or a local IDE. The main
# numerical parameters are grouped in the SimConfig dataclass below.

# ===== Fix garbled characters (keep default font family): disable TeX, increase clarity =====
mpl.rcParams['text.usetex'] = False
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['figure.dpi'] = 120

# ====================== 2. (Optional) Function to read experimental Excel data ======================
def read_experiment_excel(path, sheet_name=0):
    """
    Read experimental data and normalize column names for later plotting.
    Recognized fields: Nd, Sr/Nd (or compute from Sr & Nd), 87Sr/86Sr, 143Nd/144Nd, Label (optional).
    """
    # Read the selected Excel sheet. The function then searches for a set of
    # possible column names so that different data tables can be used without
    # manually renaming every column.
    df = pd.read_excel(path, sheet_name=sheet_name)
    aliases = {
        "Nd": ["Nd", "[Nd]", "Nd_total"],
        "Sr_over_Nd": ["Sr/Nd", "Sr_over_Nd", "SrNd", "Sr_over_Nd_ratio"],
        "R_87Sr_86Sr": ["87Sr/86Sr", "R_87Sr_86Sr", "Sr_isotope"],
        "R_143Nd_144Nd": ["143Nd/144Nd", "R_143Nd_144Nd", "Nd_isotope"],
        "Sr": ["Sr", "[Sr]", "Sr_total"],
        "Label": ["Label", "Sample", "ID", "Name"]
    }
    # Helper function: return the first matching column name from a list of
    # accepted aliases. If no column is found, return None.
    def pick(cols):
        for c in cols:
            if c in df.columns:
                return c
        return None

    out = pd.DataFrame()
    c = pick(aliases["Nd"])
    if c is not None: out["Nd"] = df[c]

    c = pick(aliases["Sr_over_Nd"])
    if c is not None:
        out["Sr_over_Nd"] = df[c]
    else:
        c_sr, c_nd = pick(aliases["Sr"]), pick(aliases["Nd"])
        if c_sr is None or c_nd is None:
            raise ValueError("Missing Sr/Nd or (Sr and Nd) columns; cannot compute Sr/Nd.")
        out["Sr_over_Nd"] = df[c_sr] / df[c_nd]

    c = pick(aliases["R_87Sr_86Sr"])
    if c is not None: out["R_87Sr_86Sr"] = df[c]

    c = pick(aliases["R_143Nd_144Nd"])
    if c is not None: out["R_143Nd_144Nd"] = df[c]

    c = pick(aliases["Label"])
    if c is not None: out["Label"] = df[c].astype(str)

    return out.dropna(how="all")


def read_grouped_experiment_excel(path, sheet_name=0):
    """
    Read one Excel file containing both MI and WR data.

    Recommended format:
      Type | Nd | Sr/Nd | 87Sr/86Sr | 143Nd/144Nd
      MI   | .. |  ..   |    ..     |     ..
      WR   | .. |  ..   |    ..     |     ..

    The Type column is used to split MI and WR.
    Fallback: if the workbook has separate sheets named MI and WR, those sheets are read.
    """
    type_aliases = ["Type", "type", "Group", "group", "SampleType", "Sample_Type", "Category", "category"]

    def pick_type_col(df):
        for c in type_aliases:
            if c in df.columns:
                return c
        return None

    # Recommended single-sheet format with Type column
    try:
        df_raw = pd.read_excel(path, sheet_name=sheet_name)
        type_col = pick_type_col(df_raw)
        if type_col is not None:
            mi_raw = df_raw[df_raw[type_col].astype(str).str.upper().str.strip() == "MI"].copy()
            wr_raw = df_raw[df_raw[type_col].astype(str).str.upper().str.strip() == "WR"].copy()

            tmp_mi = os.path.join(CFG.outdir, "__tmp_MI_group.xlsx")
            tmp_wr = os.path.join(CFG.outdir, "__tmp_WR_group.xlsx")
            mi_raw.to_excel(tmp_mi, index=False)
            wr_raw.to_excel(tmp_wr, index=False)
            mi_df = read_experiment_excel(tmp_mi)
            wr_df = read_experiment_excel(tmp_wr)
            try:
                os.remove(tmp_mi)
                os.remove(tmp_wr)
            except Exception:
                pass
            return mi_df, wr_df
    except Exception as e:
        print(f"[Warning] Failed to read grouped single-sheet Excel: {e}")

    # Fallback: separate sheets named MI and WR
    try:
        xls = pd.ExcelFile(path)
        sheet_map = {s.upper().strip(): s for s in xls.sheet_names}
        if "MI" in sheet_map and "WR" in sheet_map:
            mi_df = read_experiment_excel(path, sheet_name=sheet_map["MI"])
            wr_df = read_experiment_excel(path, sheet_name=sheet_map["WR"])
            return mi_df, wr_df
    except Exception as e:
        print(f"[Warning] Failed to read MI/WR sheets: {e}")

    raise ValueError("Excel must contain a Type column with MI/WR values, or separate sheets named MI and WR.")

# ====================== 3. Dataclass parameter structures ======================
@dataclass
class Endmember:
    """Endmember parameters: only chemical concentrations and isotope ratios."""
    Sr_total: float
    Nd_total: float
    R_87Sr_86Sr: float
    R_143Nd_144Nd: float

@dataclass
class SimConfig:
    """Global simulation configuration: A0 = constant velocity amplitude."""
    # Grid and time
    N: int = 96
    L: float = 10.0           # m
    dt: float = 1.0e7         # s  (~0.317 yr)
    steps: int = 3000         # Scatter plots use steps=1000, 1000 for tholeiitic, 3000 for alkali

    # Sine-flow parameters (constant amplitude)
    A0: float = 5e-10          # m/s
    switch_period: int = 10   # Switch flow direction every N steps (x <-> y)

    # Diffusion coefficients m2/s
    kappa_87Sr: float = 6e-11
    kappa_86Sr: float = 6e-11
    kappa_143Nd: float = 2e-11
    kappa_144Nd: float = 2e-11
    kappa_tracer: float = 6e-11 #6e-11

    # Output
    snapshot_count: int = 6
    save_npz: bool = True
    save_csv_timeseries: bool = True
    outdir: str = "mixing"

    # Time unit
    seconds_per_year: float = 365.25 * 86400.0
    time_unit_label: str = "yr"

# Endmembers used in the initial condition.
# A is placed on the left half of the domain and B is placed on the right half.
# In this example, the two endmembers have the same Sr/Nd ratio but different
# absolute Sr and Nd concentrations, which is useful for illustrating
# isotope--element decoupling during non-equilibrium mixing.
A = Endmember(30, 30, 0.7035, 0.51295)   # Sr total, Nd total, 87Sr/86Sr, 143Nd/144Nd
B = Endmember(1,  1,  0.701, 0.5138)
CFG = SimConfig()

def steps_to_years(steps):
    """Convert step number → years; used to show time in plots."""
    return np.asarray(steps, dtype=float) * CFG.dt / CFG.seconds_per_year

# ====================== 4. Grid and initial conditions ======================
N, L, dt = CFG.N, CFG.L, CFG.dt

# Grid spacing. The computational domain is a square box of side length L.
# Cell centres are used rather than cell edges.
dx = L / N
dy = L / N
x = (np.arange(N) + 0.5) * dx
y = (np.arange(N) + 0.5) * dy

# Two-dimensional coordinate arrays. The indexing="ij" convention makes the
# first array index correspond to x and the second array index correspond to y.
X, Y = np.meshgrid(x, y, indexing="ij")

def split_field(valA, valB, frac=0.5):
    """Initialize left = A, right = B."""
    F = np.zeros((N, N))
    F[X < frac*L]  = valA
    F[X >= frac*L] = valB
    return F

def split_isotopes(total_A, total_B, R_A, R_B):
    """
    Split isotopic components according to total concentration and isotope ratio:
      heavy = R/(1+R)*total
      light = 1/(1+R)*total
    Then apply left-right initialization.
    """
    heavy = split_field(R_A/(1+R_A)*total_A, R_B/(1+R_B)*total_B)
    light = split_field(1/(1+R_A)*total_A,  1/(1+R_B)*total_B)
    return heavy, light

# Initialize isotope-component fields. Instead of directly advecting isotope
# ratios, the model advects and diffuses isotope abundances/components. Isotope
# ratios are calculated from these components only when needed for output. This
# avoids non-conservative behaviour that would occur if isotope ratios were
# transported directly.
Sr87, Sr86   = split_isotopes(A.Sr_total,  B.Sr_total,  A.R_87Sr_86Sr,  B.R_87Sr_86Sr)
Nd143, Nd144 = split_isotopes(A.Nd_total,  B.Nd_total,  A.R_143Nd_144Nd, B.R_143Nd_144Nd)

# F_A is an optional passive tracer-like field indicating the initial fraction
# of endmember A. It is useful for plotting and for identifying the initial
# endmember positions.
F_A = split_field(1.0, 0.0)

# ====================== 5. Core numerics: Advection + Diffusion ======================
def semi_lagrangian_advect(C, u, v, dt_local):
    """
    Advect a scalar field using a semi-Lagrangian back-trajectory method.

    For every grid cell at the current time, the method traces a trajectory
    backward by one time step using the local velocity field. The scalar value
    at the departure point is then estimated by bilinear interpolation from the
    previous grid. Periodic boundary conditions are imposed using the modulo
    operator.
    """
    Xb = (X - u * dt_local) % L
    Yb = (Y - v * dt_local) % L
    xb = Xb / dx - 0.5
    yb = Yb / dy - 0.5
    i0 = np.floor(xb).astype(int) % N; j0 = np.floor(yb).astype(int) % N
    i1 = (i0 + 1) % N;               j1 = (j0 + 1) % N
    sx = xb - np.floor(xb);          sy = yb - np.floor(yb)
    C00 = C[i0, j0]; C10 = C[i1, j0]; C01 = C[i0, j1]; C11 = C[i1, j1]
    return (1-sx)*(1-sy)*C00 + sx*(1-sy)*C10 + (1-sx)*sy*C01 + sx*sy*C11

def diffuse_fd(C, kappa, dt_local):
    """
    Apply one explicit finite-difference diffusion step.

    The diffusion term is kappa * Laplacian(C). The Laplacian is approximated
    by second-order central differences in x and y. Periodic boundaries are
    implemented with numpy.roll.
    """
    Cxx = (np.roll(C,-1,0)-2*C+np.roll(C,1,0))/dx**2
    Cyy = (np.roll(C,-1,1)-2*C+np.roll(C,1,1))/dy**2
    return C + kappa * dt_local * (Cxx + Cyy)

def element_totals():
    """Return Sr_tot, Nd_tot."""
    return Sr87 + Sr86, Nd143 + Nd144

def ratios():
    """Compute R_Sr, R_Nd, R_SrNd, Sr_tot, Nd_tot."""
    Sr_tot, Nd_tot = element_totals()
    R_Sr   = np.where(Sr86  > 0, Sr87/Sr86,      np.nan)
    R_Nd   = np.where(Nd144 > 0, Nd143/Nd144,    np.nan)
    R_SrNd = np.where(Nd_tot > 0, Sr_tot/Nd_tot, np.nan)
    return R_Sr, R_Nd, R_SrNd, Sr_tot, Nd_tot

# ====================== 6. Diffusion stability warning ======================
kappa_max = max(CFG.kappa_87Sr, CFG.kappa_86Sr, CFG.kappa_143Nd, CFG.kappa_144Nd, CFG.kappa_tracer)
dt_stable = min(dx, dy)**2 / (4*kappa_max + 1e-30)
if dt > dt_stable:
    print(f"[Warning] dt={dt} > dt_stable≈{dt_stable:.4g}; explicit diffusion may be unstable. Reduce dt or increase N.")

# ====================== 7. Time stepping (constant-amplitude sine flow) ======================
def take_snapshot(t, snaps_dict):
    """Save snapshots of: R_Sr, R_Nd, R_SrNd, Sr_tot, Nd_tot, F_A."""
    R_Sr, R_Nd, R_SrNd, Sr_tot, Nd_tot = ratios()
    snaps_dict[t] = dict(R_Sr=R_Sr.copy(), R_Nd=R_Nd.copy(),
                         R_SrNd=R_SrNd.copy(), Sr_tot=Sr_tot.copy(),
                         Nd_tot=Nd_tot.copy(), F_A=F_A.copy())

os.makedirs(CFG.outdir, exist_ok=True)
snap_ids = np.linspace(0, CFG.steps-1, CFG.snapshot_count, dtype=int)
snaps = {}
ts, mean_srnd, med_srnd, p10_srnd, p90_srnd = [], [], [], [], []

# t = 0: initial statistics and snapshot
if 0 in snap_ids:
    take_snapshot(0, snaps)
R0 = ratios()[2]; r0 = R0[np.isfinite(R0)]
ts.append(0); mean_srnd.append(float(np.mean(r0))); med_srnd.append(float(np.median(r0)))
p10_srnd.append(float(np.percentile(r0,10))); p90_srnd.append(float(np.percentile(r0,90)))

# —— Main loop ——
for t in range(1, CFG.steps):

    amp = CFG.A0

    # Alternating sine-flow directions (switch x/y every switch_period steps)
    phase = (t // CFG.switch_period) % 2
    if phase == 0:
        u = amp * np.sin(2*np.pi*(Y/L)); v = np.zeros_like(u)
    else:
        u = np.zeros_like(X);            v = amp * np.sin(2*np.pi*(X/L))

    # Semi-Lagrangian advection step. Each isotope component and the passive
    # tracer are transported by the same velocity field.
    for C in (Sr87, Sr86, Nd143, Nd144, F_A):
        C[:] = semi_lagrangian_advect(C, u, v, dt)

    # Explicit diffusion step. Different isotope systems may use different
    # diffusion coefficients, allowing Sr and Nd isotope systems to approach
    # equilibrium at different rates.
    Sr87[:] = diffuse_fd(Sr87, CFG.kappa_87Sr, dt)
    Sr86[:] = diffuse_fd(Sr86, CFG.kappa_86Sr, dt)
    Nd143[:] = diffuse_fd(Nd143, CFG.kappa_143Nd, dt)
    Nd144[:] = diffuse_fd(Nd144, CFG.kappa_144Nd, dt)
    F_A[:]   = diffuse_fd(F_A,   CFG.kappa_tracer, dt)

    # Save snapshots
    if t in snap_ids:
        take_snapshot(t, snaps)

    # Time-series statistics
    _, _, R_SrNd, _, _ = ratios()
    r = R_SrNd[np.isfinite(R_SrNd)]
    ts.append(t)
    mean_srnd.append(float(np.mean(r)))
    med_srnd.append(float(np.median(r)))
    p10_srnd.append(float(np.percentile(r, 10)))
    p90_srnd.append(float(np.percentile(r, 90)))

# ====================== 8. Save output ======================
# Save gridded snapshot data. The NPZ file can be reloaded later without
# re-running the full simulation.
if CFG.save_npz:
    np.savez(os.path.join(CFG.outdir, "snaps.npz"),
             snaps_times=np.array(sorted(snaps.keys())),
             **{f"{t}_R_Sr":snaps[t]["R_Sr"] for t in snaps},
             **{f"{t}_R_Nd":snaps[t]["R_Nd"] for t in snaps},
             **{f"{t}_R_SrNd":snaps[t]["R_SrNd"] for t in snaps},
             **{f"{t}_Sr_tot":snaps[t]["Sr_tot"] for t in snaps},
             **{f"{t}_Nd_tot":snaps[t]["Nd_tot"] for t in snaps},
             **{f"{t}_F_A":snaps[t]["F_A"] for t in snaps},
             config=np.bytes_(str(asdict(CFG))),
             A_params=np.bytes_(str(asdict(A))),
             B_params=np.bytes_(str(asdict(B))))

# Save summary statistics of Sr/Nd through time. The 10th--90th percentile
# range is useful for visualizing the degree of spatial heterogeneity.
if CFG.save_csv_timeseries:
    with open(os.path.join(CFG.outdir, "Sr_over_Nd_timeseries.csv"), "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["step","mean_Sr/Nd","median_Sr/Nd","P10","P90"])
        for row in zip(ts, mean_srnd, med_srnd, p10_srnd, p90_srnd):
            w.writerow(row)

# ====================== 9. Plotting functions ======================
def plot_maps_at_time(t, data):
    """Six-panel map + initial endmember info in lower-right; title shows integer years."""
    R_Sr, R_Nd, R_SrNd = data["R_Sr"], data["R_Nd"], data["R_SrNd"]
    Sr_tot, Nd_tot     = data["Sr_tot"], data["Nd_tot"]
    fig, axs = plt.subplots(2, 3, figsize=(12, 7), dpi=160)
    ims = []; extent = [0, L, 0, L]

    ims.append(axs[0,0].imshow(R_Sr.T,   origin="lower", cmap="viridis", extent=extent)); axs[0,0].set_title(r"$^{87}Sr/^{86}Sr$")
    ims.append(axs[0,1].imshow(R_Nd.T,   origin="lower", cmap="viridis", extent=extent)); axs[0,1].set_title(r"$^{143}Nd/^{144}Nd$")
    ims.append(axs[0,2].imshow(R_SrNd.T, origin="lower", cmap="viridis", extent=extent)); axs[0,2].set_title("Sr/Nd")

    ims.append(axs[1,0].imshow(Sr_tot.T, origin="lower", cmap="magma", extent=extent)); axs[1,0].set_title("[Sr] total")
    ims.append(axs[1,1].imshow(Nd_tot.T, origin="lower", cmap="magma", extent=extent)); axs[1,1].set_title("[Nd] total")

    # Right-lower information panel (no viscosity)
    ax_info = axs[1,2]; ax_info.axis("off"); ax_info.set_title("Initial endmembers (t = 0)")
    iso_sr = r"$^{87}\mathrm{Sr}/^{86}\mathrm{Sr}$"; iso_nd = r"$^{143}\mathrm{Nd}/^{144}\mathrm{Nd}$"
    left_text = "\n".join([
        "Left = A",
        f"[Sr] total : {A.Sr_total:.6g}",
        f"[Nd] total : {A.Nd_total:.6g}",
        f"{iso_sr}  : {A.R_87Sr_86Sr:.6g}",
        f"{iso_nd}  : {A.R_143Nd_144Nd:.6g}",
    ])
    right_text = "\n".join([
        "Right = B",
        f"[Sr] total : {B.Sr_total:.6g}",
        f"[Nd] total : {B.Nd_total:.6g}",
        f"{iso_sr}  : {B.R_87Sr_86Sr:.6g}",
        f"{iso_nd}  : {B.R_143Nd_144Nd:.6g}",
    ])
    ax_info.text(0.06, 0.92, left_text,  transform=ax_info.transAxes, ha='left', va='top', fontsize=9, linespacing=1.9)
    ax_info.text(0.60, 0.92, right_text, transform=ax_info.transAxes, ha='left', va='top', fontsize=9, linespacing=1.9)

    for ax in axs.ravel()[:5]:
        ax.set_xlim(0, L); ax.set_ylim(0, L)
        ax.set_xlabel("x (m)"); ax.set_ylabel("y (m)")
    for im, ax in zip(ims, axs.ravel()[:5]):
        plt.colorbar(im, ax=ax, fraction=0.046)

    Ty = steps_to_years(t)
    fig.suptitle(f"T = {Ty:.0f} {CFG.time_unit_label}  (t = {t} step)")
    plt.tight_layout()
    plt.savefig(os.path.join(CFG.outdir, f"maps_t{t}.png"), dpi=160)
    plt.show()


def plot_combined_srnd_2x2(
    snaps, times_to_plot,
    ts, mean_srnd, med_srnd, p10_srnd, p90_srnd,
    mi_df=None,
    wr_df=None,
    mi_label="MI",
    wr_label="WR",
    mi_size=80,
    wr_size=80,
    mi_marker='o',
    wr_marker='s',
    mi_color='#377eb8',
    wr_color='#e41a1c',
    exp_edgecolor='k', exp_alpha=0.6):

    """
    2×2 multi-panel figure:
      (0,0) Nd vs Sr/Nd
      (0,1) Sr/Nd time series
      (1,0) Sr/Nd vs 87Sr/86Sr
      (1,1) Sr/Nd vs 143Nd/144Nd

    Enhancement: use large star marker for T=0, rendered on top.
    """

    # Although the function name contains "2x2", this version uses a vertical
    # 4-panel layout, which is convenient for manuscript or supplement figures.
    fig, axs = plt.subplots(4, 1, figsize=(7, 22), dpi=150)

    # Add panel labels (a)--(d). Change this list if the panel order is modified.
    panel_labels = ['(a)', '(b)', '(c)', '(d)']
    for label, ax in zip(panel_labels, axs.ravel()):
        ax.text(0.98, 0.97, label, transform=ax.transAxes,
            fontsize=14, fontweight='bold',
            ha='right', va='top')

    # Sorting: non-zero times first, T=0 last to ensure overlay order
    tp = list(times_to_plot)
    t0_present = (0 in tp)
    others = [t for t in tp if t != 0]
    ordered_times = others + ([0] if t0_present else [])

    # Style for T=0
    # Endmember styles
    A_STYLE = dict(
        s=180,
        marker='*',
        c='gold',
        edgecolors='k',
        linewidths=1.0,
        alpha=0.95,
        zorder=7
    )

    B_STYLE = dict(
        s=180,
        marker='*',
        c='deepskyblue',
        edgecolors='k',
        linewidths=1.0,
        alpha=0.95,
        zorder=7
    )

    # Style for other times
    OTHER_STYLE = dict(s=6, alpha=0.2, zorder=3)  # Style for model particles at non-zero times; lower alpha gives more transparent points.

    # =======================================
    # (0,0) Nd vs Sr/Nd
    # =======================================
    ax = axs[0]

    for t in others:
        Nd   = snaps[t]["Nd_tot"].ravel()
        SrNd = (snaps[t]["Sr_tot"] / snaps[t]["Nd_tot"]).ravel()
        idx  = np.random.choice(Nd.size, size=min(4000, Nd.size), replace=False)
        Ty   = steps_to_years(t)
        ax.scatter(Nd[idx], SrNd[idx],
                   label=f"T={Ty:.0f} {CFG.time_unit_label}",
                   **OTHER_STYLE)

    if t0_present:
        t = 0
        Nd   = snaps[t]["Nd_tot"].ravel()
        SrNd = (snaps[t]["Sr_tot"] / snaps[t]["Nd_tot"]).ravel()
        FA   = snaps[t]["F_A"].ravel()

        A_mask = FA > 0.5
        B_mask = FA <= 0.5

        ax.scatter(
            Nd[A_mask], SrNd[A_mask],
            label="Endmember A",
            **A_STYLE
        )

        ax.scatter(
            Nd[B_mask], SrNd[B_mask],
            label="Endmember B",
            **B_STYLE
        )

    if mi_df is not None and {"Nd","Sr_over_Nd"}.issubset(mi_df.columns):
        ax.scatter(mi_df["Nd"], mi_df["Sr_over_Nd"],
                   s=mi_size, marker=mi_marker, c=mi_color,
                   edgecolors=exp_edgecolor, linewidths=0.8, alpha=exp_alpha,
                   label=mi_label, zorder=4)

    if wr_df is not None and {"Nd","Sr_over_Nd"}.issubset(wr_df.columns):
        ax.scatter(wr_df["Nd"], wr_df["Sr_over_Nd"],
                   s=wr_size, marker=wr_marker, c=wr_color,
                   edgecolors=exp_edgecolor, linewidths=0.8, alpha=exp_alpha,
                   label=wr_label, zorder=4)

    ax.set_xlabel("Nd-N");
    ax.set_ylabel("Sr/Sr*")
    ax.grid(False)
    ax.legend(loc='upper center',
              bbox_to_anchor=(0.8, 1),  # Legend position: first value controls left--right shift; second controls up--down shift.
              ncol=1, frameon=False)

    # =======================================
    # (0,1) Sr/Nd time series
    # =======================================
    ax = axs[1]
    t_years = steps_to_years(ts)
    line_mean,   = ax.plot(t_years, mean_srnd, linewidth=2, label="Mean")
    line_median, = ax.plot(t_years, med_srnd, linestyle="--", label="Median")
    band = ax.fill_between(t_years, p10_srnd, p90_srnd, alpha=0.2)
    ax.set_xlim(0, 1000)  # Set the time-axis range for this panel.
    band_color = band.get_facecolor()[0]

    from matplotlib.lines import Line2D
    band_handle = Line2D([0], [0], lw=10, alpha=0.25,
                         color=band_color, label="10th–90th percentile range")

    ax.set_xlabel(f"Time ({CFG.time_unit_label})")
    ax.set_ylabel("Sr/Sr*")
    ax.grid(False)
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    ax.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
    ax.legend(handles=[line_mean, line_median, band_handle],
              frameon=False, loc="upper center")

    # =======================================
    # (1,0) Sr/Nd vs 87Sr/86Sr
    # =======================================
    ax = axs[2]
    for t in others:
        R_Sr   = snaps[t]["R_Sr"].ravel()
        SrNd   = snaps[t]["R_SrNd"].ravel()
        m = np.isfinite(R_Sr) & np.isfinite(SrNd)
        if m.any():
            idx = np.random.choice(np.where(m)[0],
                                   size=min(4000, m.sum()),
                                   replace=False)
            Ty = steps_to_years(t)
            ax.scatter(R_Sr[idx], SrNd[idx],
                       label=f"T={Ty:.0f} {CFG.time_unit_label}",
                       **OTHER_STYLE)

    if t0_present:
        t = 0
        R_Sr = snaps[t]["R_Sr"].ravel()
        SrNd = snaps[t]["R_SrNd"].ravel()
        FA   = snaps[t]["F_A"].ravel()

        A_mask = FA > 0.5
        B_mask = FA <= 0.5

        ax.scatter(
            R_Sr[A_mask], SrNd[A_mask],
            label="Endmember A",
            **A_STYLE
        )

        ax.scatter(
            R_Sr[B_mask], SrNd[B_mask],
            label="Endmember B",
            **B_STYLE
        )

    if mi_df is not None and {"Sr_over_Nd","R_87Sr_86Sr"}.issubset(mi_df.columns):
        ax.scatter(mi_df["R_87Sr_86Sr"], mi_df["Sr_over_Nd"],
                   s=mi_size, marker=mi_marker, c=mi_color,
                   edgecolors=exp_edgecolor, linewidths=0.8, alpha=exp_alpha,
                   label=mi_label, zorder=4)

    if wr_df is not None and {"Sr_over_Nd","R_87Sr_86Sr"}.issubset(wr_df.columns):
        ax.scatter(wr_df["R_87Sr_86Sr"], wr_df["Sr_over_Nd"],
                   s=wr_size, marker=wr_marker, c=wr_color,
                   edgecolors=exp_edgecolor, linewidths=0.8, alpha=exp_alpha,
                   label=wr_label, zorder=4)

    ax.set_xlabel(r"$^{87}\mathrm{Sr}/^{86}\mathrm{Sr}$")
    ax.set_ylabel("Sr/Sr*")
    ax.grid(False)
    ax.legend(frameon=False)
    if ax.get_legend(): ax.get_legend().remove()

    # =======================================
    # (1,1) Sr/Nd vs 143Nd/144Nd
    # =======================================
    ax = axs[3]
    for t in others:
        R_Nd   = snaps[t]["R_Nd"].ravel()
        SrNd   = snaps[t]["R_SrNd"].ravel()
        m = np.isfinite(R_Nd) & np.isfinite(SrNd)
        if m.any():
            idx = np.random.choice(np.where(m)[0],
                                   size=min(4000, m.sum()),
                                   replace=False)
            Ty = steps_to_years(t)
            ax.scatter(R_Nd[idx], SrNd[idx],
                       label=f"T={Ty:.0f} {CFG.time_unit_label}",
                       **OTHER_STYLE)

    if t0_present:
        t = 0
        R_Nd = snaps[t]["R_Nd"].ravel()
        SrNd = snaps[t]["R_SrNd"].ravel()
        FA   = snaps[t]["F_A"].ravel()

        A_mask = FA > 0.5
        B_mask = FA <= 0.5

        ax.scatter(
            R_Nd[A_mask], SrNd[A_mask],
            label="Endmember A",
            **A_STYLE
        )

        ax.scatter(
            R_Nd[B_mask], SrNd[B_mask],
            label="Endmember B",
            **B_STYLE
        )

    if mi_df is not None and {"Sr_over_Nd","R_143Nd_144Nd"}.issubset(mi_df.columns):
        ax.scatter(mi_df["R_143Nd_144Nd"], mi_df["Sr_over_Nd"],
                   s=mi_size, marker=mi_marker, c=mi_color,
                   edgecolors=exp_edgecolor, linewidths=0.8,
                   alpha=exp_alpha, label=mi_label, zorder=4)

    if wr_df is not None and {"Sr_over_Nd","R_143Nd_144Nd"}.issubset(wr_df.columns):
        ax.scatter(wr_df["R_143Nd_144Nd"], wr_df["Sr_over_Nd"],
                   s=wr_size, marker=wr_marker, c=wr_color,
                   edgecolors=exp_edgecolor, linewidths=0.8,
                   alpha=exp_alpha, label=wr_label, zorder=4)

    ax.set_xlabel(r"$^{143}\mathrm{Nd}/^{144}\mathrm{Nd}$")
    ax.set_ylabel("Sr/Sr*")
    ax.grid(False)
    ax.legend(frameon=False, loc="upper left")
    if ax.get_legend(): ax.get_legend().remove()

    fig.suptitle("Iceland",
             fontsize=22,
             fontweight='bold',
             y=0.955)
    plt.tight_layout(rect=[0,0,1,0.965])
    plt.savefig(os.path.join(CFG.outdir, "Combined_SrNd_2x2.png"), dpi=160)
    plt.show()

# Sr--Nd isotope cross-plot. This figure helps visualize whether the two
# isotope systems remain coupled or become partially decoupled during mixing.
def plot_sr_nd_isotope_evolution(
    snaps,
    times_to_plot,
    mi_df=None,
    wr_df=None,
    mi_label="MI",
    wr_label="WR",
    mi_size=80,
    wr_size=80,
    mi_marker='o',
    wr_marker='s',
    mi_color='gray',
    wr_color='#0072B2',
    exp_edgecolor='k',
    exp_alpha=0.6,
    outname="Sr_Nd_isotope_evolution.png"
):
    """
    Plot 87Sr/86Sr vs 143Nd/144Nd evolution through time.
    x-axis = Sr isotope
    y-axis = Nd isotope
    MI and WR data can be overlaid.
    """

    fig, ax = plt.subplots(figsize=(7, 6), dpi=150)

    # plot simulation snapshots
    for t in times_to_plot:
        R_Sr = snaps[t]["R_Sr"].ravel()
        R_Nd = snaps[t]["R_Nd"].ravel()

        m = np.isfinite(R_Sr) & np.isfinite(R_Nd)
        if m.any():
            idx = np.random.choice(
                np.where(m)[0],
                size=min(4000, m.sum()),
                replace=False
            )

            Ty = steps_to_years(t)

            ax.scatter(
                R_Sr[idx],
                R_Nd[idx],
                s=8,
                alpha=0.45,
                label=f"T={Ty:.0f} {CFG.time_unit_label}"
            )

    # overlay MI data
    if mi_df is not None and {"R_87Sr_86Sr", "R_143Nd_144Nd"}.issubset(mi_df.columns):
        ax.scatter(
            mi_df["R_87Sr_86Sr"],
            mi_df["R_143Nd_144Nd"],
            s=mi_size,
            marker=mi_marker,
            facecolors='none',
            edgecolors=mi_color,
            linewidths=1.2,
            alpha=exp_alpha,
            label=mi_label,
            zorder=5
        )

    # overlay WR data
    if wr_df is not None and {"R_87Sr_86Sr", "R_143Nd_144Nd"}.issubset(wr_df.columns):
        ax.scatter(
            wr_df["R_87Sr_86Sr"],
            wr_df["R_143Nd_144Nd"],
            s=wr_size,
            marker=wr_marker,
            facecolors='none',
            edgecolors=wr_color,
            linewidths=1.2,
            alpha=exp_alpha,
            label=wr_label,
            zorder=5
        )

    ax.set_xlabel(r"$^{87}\mathrm{Sr}/^{86}\mathrm{Sr}$")
    ax.set_ylabel(r"$^{143}\mathrm{Nd}/^{144}\mathrm{Nd}$")

    ax.grid(False)
    ax.legend(frameon=False, loc="best")

    plt.tight_layout()
    plt.savefig(os.path.join(CFG.outdir, outname), dpi=160)
    plt.show()

    print(f"[Info] Saved Sr-Nd isotope evolution figure: {outname}")

# ====================== 12. New: 2×3 Sr/Nd field maps ======================
def plot_srnd_fields_2x3_with_annotation(snaps, steps, outname="SrNd_fields_2x3_annotated.png"):
    """
    Plot 6 Sr/Nd field snapshots in a 2×3 layout.
    Annotation and division line added to the t=0 panel.
    Automatic times: 0, 1/5, 2/5, 3/5, 4/5, final.
    """
    # Select 6 time steps from 0 → steps-1
    times_to_plot = np.linspace(0, steps - 1, 6, dtype=int)
    times_to_plot = [t for t in times_to_plot if t in snaps]

    # Unified colormap range (exclude NaN)
    all_vals = []
    for t in times_to_plot:
        SrNd = snaps[t]["R_SrNd"]
        all_vals.append(SrNd[np.isfinite(SrNd)])
    vmin = min([v.min() for v in all_vals])
    vmax = max([v.max() for v in all_vals])

    # Plot 2×3 figure
    fig, axs = plt.subplots(2, 3, figsize=(14, 8.5), dpi=150)
    extent = [0, CFG.L, 0, CFG.L]

    for ax, t in zip(axs.ravel(), times_to_plot):
        SrNd = snaps[t]["R_SrNd"].T
        im = ax.imshow(SrNd, origin="lower", extent=extent,
                       cmap="viridis", vmin=vmin, vmax=vmax)
        Ty = steps_to_years(t)
        ax.set_title(f"T = {Ty:.0f} {CFG.time_unit_label}", fontsize=16)
        ax.set_xlabel("x (m)", fontsize=14)
        ax.set_ylabel("y (m)", fontsize=14)
        ax.tick_params(axis='both', labelsize=12)
        plt.colorbar(im, ax=ax, fraction=0.046)

        # Extra annotation for t = 0
        if t == 0:
            ax.axvline(CFG.L / 2, color='white', linestyle='--', linewidth=1.5)

            left_text = "\n".join([
                "Left = A",
                f"Sr-N : {A.Sr_total:.6g}",
                f"Nd-N : {A.Nd_total:.6g}"
            ])
            right_text = "\n".join([
                "Right = B",
                f"Sr-N : {B.Sr_total:.6g}",
                f"Nd-N : {B.Nd_total:.6g}"
            ])
            ax.text(0.3, CFG.L - 0.4, left_text,
                    ha='left', va='top', fontsize=14, color='white')
            ax.text(CFG.L/2 + 0.3, CFG.L - 0.4, right_text,
                    ha='left', va='top', fontsize=14, color='white')

    fig.suptitle("Sr/Sr* Field Evolution", fontsize=20, y=1.02)
    plt.tight_layout()
    plt.savefig(os.path.join(CFG.outdir, outname), dpi=160)
    plt.show()
    print(f"[Info] Saved 2×3 Sr/Nd annotated field map: {outname}")

# ====================== 10. Output and experimental-data overlay ======================
# The code below runs after the simulation is complete. It first exports model
# maps for the selected snapshot times, then optionally overlays natural MI/WR
# data if a compatible Excel file is available.
print(f"[Info] Output directory: {os.path.abspath(CFG.outdir)}")
print(f"[Info] dx={dx:.6g} m, dt={dt:.6g} s ≈ {dt/CFG.seconds_per_year:.3f} {CFG.time_unit_label}; A0={CFG.A0:.3e} m/s")

for t in snap_ids:
    plot_maps_at_time(t, snaps[t])

# Optional: load MI and WR data from one Excel file
# Recommended Excel format:
#   Type | Nd | Sr/Nd | 87Sr/86Sr | 143Nd/144Nd
#   MI   | .. |  ..   |    ..     |     ..
#   WR   | .. |  ..   |    ..     |     ..
#
# Fallback: the same file may also contain two sheets named "MI" and "WR".
EXP_PATH = "MI_WR.xlsx"
EXP_SHEET = "Sheet1"

mi_df = None
wr_df = None

try:
    if os.path.exists(EXP_PATH):
        mi_df, wr_df = read_grouped_experiment_excel(EXP_PATH, sheet_name=EXP_SHEET)
        print(f"[Info] Loaded grouped MI/WR data: {EXP_PATH} / {EXP_SHEET}")
        print(f"[Info] MI rows={len(mi_df)}, WR rows={len(wr_df)}")
    else:
        print(f"[Note] Grouped Excel not found: {EXP_PATH} (simulation only)")
except Exception as e:
    print(f"[Warning] Failed to parse grouped MI/WR Excel: {e} (simulation only)")

plot_combined_srnd_2x2(
    snaps, snap_ids,
    ts, mean_srnd, med_srnd, p10_srnd, p90_srnd,
    mi_df=mi_df,
    wr_df=wr_df,
    mi_label="MI",
    wr_label="WR",
    mi_size=80,
    wr_size=80,
    mi_marker='o',
    wr_marker='s',
    mi_color='#377eb8',
    wr_color='#e41a1c',
    exp_edgecolor='k', exp_alpha=1  # Transparency of the natural MI and WR data points.
)

plot_srnd_fields_2x3_with_annotation(snaps, CFG.steps)

plot_sr_nd_isotope_evolution(
    snaps,
    snap_ids,
    mi_df=mi_df,
    wr_df=wr_df,
    mi_label="MI",
    wr_label="WR",
    mi_size=80,
    wr_size=80,
    mi_marker='o',
    wr_marker='s',
    mi_color='#377eb8',
    wr_color='#e41a1c',
    exp_alpha=0.7
)
```
